# seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Line Plot

```python
#matplotlib
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]
plt.plot(var,var_1)
plt.show()
```

```python
#seaborn
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]
x=pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x)
plt.show()
```

```python
y=sns.load_dataset("penguins").head(20)
y
#.head(20) likha hei for first 20 records ke liye 
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex",style="sex",
             palette="Accent",markers=["o",">"],dashes=False,legend=False) 
plt.grid()
plt.title("Python",fontsize=15)
plt.show()
#hue adds label automatically 
#style changes the format of line
#palette helps to change the colour of the graphs 
#marker highlights the point 
#dashes removes the dotted line 
#yaha(in seaborn) legend hatane ke liye legend likhna padta hei 
```

## Histogram

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color="g")
plt.show()
#bins is used to organize the data
#kde - kernel density estimator - to draw a line connecting points 
#rug - color density show kar sakte hei at the bottom
```

# Count Plot

```python
var=sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette="bwr",saturation=0.8)
plt.show()
#to make the count plot HORIZONTALLY -> sns.countplot(y="sex",data=var,hue="smoker")
```

```python
#Just to see diffrence b/w bar plot and count plot
sns.barplot(x="sex",y="size",data=var)
plt.show()
```
