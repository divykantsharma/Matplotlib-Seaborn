```python
import numpy as np
import matplotlib.pyplot as plt 
```

# Bar Plot

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
c=["r","magenta","b","green"]
plt.xlabel("language",fontsize=12)
plt.ylabel("number",fontsize=12)
plt.title("Bar Plot")
plt.bar(x,y,width=0.8,color=c,align="center")
plt.show()
```

# Scatter Plot

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

# Adding text

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]
plt.plot(x,y)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)
plt.title("Labelled visual",fontsize=20)

plt.text(2,5,"text",fontsize=15,style="italic",bbox={"facecolor":"r"})
plt.annotate("python",xy=(2,1),xytext=(3,3),arrowprops=dict(facecolor="black",shrink=100))

plt.show()
```
